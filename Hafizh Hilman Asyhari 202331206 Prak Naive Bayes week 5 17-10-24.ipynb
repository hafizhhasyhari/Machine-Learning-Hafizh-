```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('kaca.csv', sep = ';')
dataset
```

```python
dataset.info()
```

```python
categorical_feature = [feature for feature in dataset.columns
                       if dataset [feature].dtypes == "O"]
dataset[categorical_feature]
```

```python
for i in categorical_feature:
    print(dataset[i].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for feature in categorical_feature:
    print(dataset[feature].unique())
    dataset[feature] = label_encoder.fit_transform(dataset[feature])
    print(dataset[feature].unique())
```

```python
dataset
```

```python
x = dataset.drop(['Kelas','No Data'], axis = 1)
y = dataset['Kelas']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit (x_train, y_train)
```

```python
x_pred_train = model.predict(x_train)
x_pred_test = model.predict(x_test)
```
